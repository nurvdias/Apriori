### To perform the apriori algorithm, we first have to break up the IDs into readable format

```python
import re

def getGenres(item):
    genre_groups = item.split(',')
    regEx = re.compile('[A-Za-z]+')
    genre = []    
    
    for group in genre_groups:
        idName = group.split(':')
        #print('idName: ', idName)
        name = idName[1].strip().replace('}','')
        if(regEx.match(name)):
            genre.append(name)
    return genre
```

```python
df.genres = df.genres.str.replace('[','').str.replace(']','').str.replace(' ','').str.replace('\"','')
df.genres
completeList = []

for item in df.genres:
    if(item != ''):
        local_genres = getGenres(item)
        completeList.append(local_genres)
```

```python
completeList[:10]
```

#### The IDs are in readable format for the Apriori algorithm

```python
import sys
sys.path.insert(0, 'Apriori-master')

from apriori import apriori
```

```python
minimum_support = 0.01
minimum_confidence = 0.01

ap = apriori.Apriori(completeList,minimum_support,minimum_confidence)
ap.run()
ap.print_frequent_itemset()
```
